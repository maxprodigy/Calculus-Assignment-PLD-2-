# Task 1: Write a dynamic function to find the derivative of any function f(x)

````python
import sympy as sp

# Create a function that calculates the derivative for any function
def dy_dx(function):
    # Define the variable x
    x = sp.symbols('x')

    # Convert the input function string into a sympy expression
    f_x = sp.sympify(function)

    # Compute the derivative of the function f(x) with respect to x
    derivative = sp.diff(f_x, x)

    # Return the derivative
    return derivative
````

````python
# Import the sympy library for symbolic mathematics.
import sympy as sp

# Function to classify critical points of a given function.
def classify_critical_points(func):
    x = sp.symbols('x')  # Define a symbolic variable 'x'.
    f_x = sp.sympify(func)  # Convert function string to a sympy expression.

    first_derivative = sp.diff(f_x, x)  # Compute the first derivative.
    second_derivative = sp.diff(first_derivative, x)  # Compute the second derivative.

    critical_points = sp.solve(first_derivative, x)  # Find critical points.

    # Classify points based on the second derivative.
    min_points = [p.evalf() for p in critical_points if second_derivative.subs(x, p) > 0]
    max_points = [p.evalf() for p in critical_points if second_derivative.subs(x, p) < 0]

    return min_points, max_points, first_derivative

# Define the function to analyze.
function_string = "x**5 - 10*x**4 + 35*x**3 - 50*x**2 + 24*x"

# Call the function to classify critical points.
min_points, max_points, first_derivative = classify_critical_points(function_string)

# Output the first derivative and critical points.
print("First Derivative:", first_derivative)
for point in min_points[:2]:  # Print minimum points.
    print(f"\nMinimum at x = {point:.2f}: f'({point:.2f}) = {first_derivative.subs(sp.symbols('x'), point):.2f}")

for point in max_points[:2]:  # Print maximum points.
    print(f"\nMaximum at x = {point:.2f}: f'({point:.2f}) = {first_derivative.subs(sp.symbols('x'), point):.2f}")
````

````python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.Symbol('x')
f = x**5 - 10*x**4 + 35*x**3 - 50*x**2 + 24*x

# Find critical points
critical_points = sp.solve(sp.diff(f, x), x)

# Classify points
minima = []
maxima = []
for point in critical_points:
    if sp.diff(f, x, 2).subs(x, point) > 0:
        minima.append(float(point))
    elif sp.diff(f, x, 2).subs(x, point) < 0:
        maxima.append(float(point))

print("Minima x-values:", minima)
print("Maxima x-values:", maxima)

# Compute global minima and maxima
global_min = min([(point, float(f.subs(x, point))) for point in critical_points], key=lambda p: p[1])
global_max = max([(point, float(f.subs(x, point))) for point in critical_points], key=lambda p: p[1])

print(f"Global minima: ({float(global_min[0]):.2f}, {global_min[1]:.2f})")
print(f"Global maxima: ({float(global_max[0]):.2f}, {global_max[1]:.2f})")

# Visualize
x_range = np.linspace(0, 4, 1000)
y_range = [float(f.subs(x, xi)) for xi in x_range]

plt.figure(figsize=(12, 8))
plt.plot(x_range, y_range, label='f(x)')
plt.scatter(minima, [float(f.subs(x, m)) for m in minima], color='red', s=100, label='Local Minima')
plt.scatter(maxima, [float(f.subs(x, m)) for m in maxima], color='green', s=100, label='Local Maxima')
plt.scatter(float(global_min[0]), global_min[1], color='purple', s=200, marker='*', label='Global Minimum')
plt.scatter(float(global_max[0]), global_max[1], color='orange', s=200, marker='*', label='Global Maximum')

plt.title("Function f(x) with Minima and Maxima")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
````

# Task 5: Finding the Area under the curve

````python
import numpy as np
from scipy.integrate import quad

# Define the function to analyze
def f(x):
    return x**5 - 10*x**4 + 35*x**3 - 50*x**2 + 24*x

# Local extrema points
local_max_1 = 0.3555671318417314
local_min_1 = 1.456087744097662
local_max_2 = 2.543912255902338
local_min_2 = 3.6444328681582685

# Calculate areas under the curve
area1, _ = quad(f, local_max_1, local_min_1)  # Area between first max and first min
area2, _ = quad(f, local_min_1, local_max_2)  # Area between first min and second max
area3, _ = quad(f, local_max_2, local_min_2)  # Area between second max and second min

# Print the absolute areas calculated
print(f"Area under the curve from {local_max_1:.2f} to {local_min_1:.2f}: {abs(area1):.4f}")
print(f"Area under the curve from {local_min_1:.2f} to {local_max_2:.2f}: {abs(area2):.4f}")
print(f"Area under the curve from {local_max_2:.2f} to {local_min_2:.2f}: {abs(area3):.4f}")

# Total area
total_area = abs(area1) + abs(area2) + abs(area3)
print(f"\nTotal area under the curve: {total_area:.4f}")
````

#Finally, What is the global Minima and the Global maxima _ Plot this so that I can see

finding the global minima and maxima of the function allows you to analyze the function over the specified range and then plot it, highlighting the critical points.

````python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Define the function
def f(x):
    return x**5 - 10*x**4 + 35*x**3 - 50*x**2 + 24*x

# Generate x values
x = np.linspace(-1, 5, 400)
y = f(x)

# Find critical points using sympy
x_sym = sp.symbols('x')
f_sym = x_sym**5 - 10*x_sym**4 + 35*x_sym**3 - 50*x_sym**2 + 24*x_sym
first_derivative = sp.diff(f_sym, x_sym)
critical_points = sp.solve(first_derivative, x_sym)

# Evaluate function at critical points
critical_values = [(float(p), f(float(p))) for p in critical_points]

# Identify global minima and maxima
global_min = min(critical_values, key=lambda point: point[1])
global_max = max(critical_values, key=lambda point: point[1])

# Plot the function
plt.figure(figsize=(12, 8))
plt.plot(x, y, label='f(x)', color='blue')

# Plot global minima and maxima
plt.plot(global_min[0], global_min[1], 'ro', label='Global Minima')
plt.plot(global_max[0], global_max[1], 'go', label='Global Maxima')

# Annotate points
plt.annotate(f'Min: {global_min[1]:.2f} at x = {global_min[0]:.2f}',
             xy=(global_min[0], global_min[1]),
             xytext=(global_min[0] + 0.5, global_min[1] + 10),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.annotate(f'Max: {global_max[1]:.2f} at x = {global_max[0]:.2f}',
             xy=(global_max[0], global_max[1]),
             xytext=(global_max[0] + 0.5, global_max[1] - 20),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.title("Function f(x) and its Global Minima and Maxima")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# Output global minima and maxima
print(f"Global Minima: {global_min}")
print(f"Global Maxima: {global_max}")
````
